```python
import pandas as pd

performance = pd.read_csv("data/raw/07_scheme_performance.csv")

print("Missing Values:")
print(performance.isnull().sum())


print("\nDuplicate Rows:", performance.duplicated().sum())

valid_expense = performance[(performance["expense_ratio_pct"] >= 0.1) &(performance["expense_ratio_pct"] <= 2.5)]

print("\nValid Expense Ratio Records:", len(valid_expense))

negative_sharpe = performance[performance["sharpe_ratio"] < 0]

print("Negative Sharpe Ratio Records:", len(negative_sharpe))
```

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("sqlite:///bluestock_mf.db")
```

```python
fund = pd.read_csv("data/processed/clean_fund_master.csv")
nav = pd.read_csv("data/processed/clean_nav.csv")
aum = pd.read_csv("data/processed/clean_aum.csv")
sip = pd.read_csv("data/processed/clean_sip_inflows.csv")
category = pd.read_csv("data/processed/clean_category_inflows.csv")
folio = pd.read_csv("data/processed/clean_folio_count.csv")
performance = pd.read_csv("data/processed/clean_performance.csv")
transactions = pd.read_csv("data/processed/clean_transactions.csv")
holdings = pd.read_csv("data/processed/clean_portfolio_holdings.csv")
benchmark = pd.read_csv("data/processed/clean_benchmark.csv")
```

```python
fund.to_sql("dim_fund", engine, if_exists="replace", index=False)

nav.to_sql("fact_nav", engine, if_exists="replace", index=False)

aum.to_sql("fact_aum", engine, if_exists="replace", index=False)

sip.to_sql("fact_sip", engine, if_exists="replace", index=False)

category.to_sql("fact_category", engine, if_exists="replace", index=False)

folio.to_sql("fact_folio", engine, if_exists="replace", index=False)

performance.to_sql("fact_performance", engine, if_exists="replace", index=False)

transactions.to_sql("fact_transactions", engine, if_exists="replace", index=False)

holdings.to_sql("fact_holdings", engine, if_exists="replace", index=False)

benchmark.to_sql("fact_benchmark", engine, if_exists="replace", index=False)

print("All datasets loaded successfully!")
```
